```python
import matplotlib.pyplot as plt
from megatron.transformations import TRANSFORMATIONS
# Assuming `original_image` is your original image and `transformations` is your dictionary
# transformations = {'name1': func1, 'name2': func2, ..., 'name11': func11}
original_image = plt.imread("./../data/lenna.jpg")
fig, axes = plt.subplots(3, 4, figsize=(20, 15))  # Adjust the size as needed

# Plot original image
axes[0, 0].imshow(original_image, cmap='gray')  # Use appropriate color map for your image
axes[0, 0].set_title('Original')
axes[0, 0].axis('off')  # Hide axes

# Plot transformed images
for i, (func,gen_kwargs) in enumerate(TRANSFORMATIONS, start=1):
    kwargs = gen_kwargs()
    transformed_image = func(original_image, **kwargs)
    row, col = divmod(i, 4)
    axes[row, col].imshow(transformed_image, cmap='gray')  # Use appropriate color map
    axes[row, col].set_title(func.__name__)
    axes[row, col].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```
